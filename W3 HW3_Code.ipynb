### ùëì(ùë•) = ùë•4 ‚àí 6ùë•2 + 4ùë• + 18

```python
def f(x):
    return x**4 - 6 * (x**2) + 4*x +18

def df(x):
    return 4*(x**3) - (12*x) + 4
```

### (iv) Write your own code for ùëÜ steps of GD on this function (do not use built-in or third-party GD codes).

```python
def gradient_descent(S, alpha, x):
    x_values = {} 
    for i in range(S):
        x = x - alpha * df(x)
        step = f'x{i + 1}'
        x_values[step] = x
    return x_values

# Test for(i) S=3 with alpha=0.1 and initial_x=1.0
S = 3
alpha = 0.1
initial_x = 1.0

final_x = gradient_descent(S, alpha, initial_x)
print("Final x value after", S, "steps of Gradient Descent:", final_x)

# Test for(ii) S=3 with alpha=0.1 and initial_x=1.0
S = 3
alpha = 0.1
initial_x = 0.0

final_x = gradient_descent(S, alpha, initial_x)
print("Final x value after", S, "steps of Gradient Descent:", final_x)

```

# Problem 2
## The College.csv dataset contains admissions data for a sample of 777 universities. We want to predict the number of applications received (‚ÄòApps‚Äô) using the other variables in the dataset.

```python
df = pd.read_csv('W3_College.csv')

df['Private'][df['Private'] == 'Yes'] = 1
df['Private'][df['Private'] == 'No'] = 0

df.info()
df.head()
```

### (i) Let the first 600 observations be the training set and the remaining 177 observations be the test set.

```python
train_size = 600

train_set, test_set = df[:600], df[600:]

print("Training set size:", len(train_set))
print("Test set size:", len(test_set))

X_train = train_set.drop('Apps', axis=1)
y_train = train_set['Apps']

X_test = test_set.drop('Apps', axis=1)
y_test = test_set['Apps']

#train_set
```

### (ii) Fit the OLS regression on the training set, and report the test error obtained.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model_ols = LinearRegression().fit(X_train, y_train)

y_pred_ols = model_ols.predict(X_test)
test_error_ols = mean_squared_error(y_test, y_pred_ols)

print("OLS Regression Test error:", test_error_ols)
```

### For the rest of the problem, let the penalization parameter vary on the 1000-point grid from 0.01 to 60.

### (iii) Fit the LASSO regression on the training set, with the penalization parameter chosen by 20-fold cross-validation. Report the test error obtained.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_private = X_train['Private']
X_test_private = X_test['Private']

X_train_scaled = scaler.fit_transform(X_train.drop('Private', axis=1))
X_test_scaled = scaler.transform(X_test.drop('Private', axis=1))

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.drop('Private', axis=1).columns)
X_train_scaled['Private'] = X_train_private

X_test = X_test.reset_index(drop=True)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.drop('Private', axis=1).columns)
X_test_scaled['Private'] = X_test_private


X_train_scaled.info()
X_test_scaled.info()
```

```python
from sklearn.linear_model import LassoCV

alphas_lasso = np.linspace(0.01, 60, 1000)

model_lasso = LassoCV(alphas = alphas_lasso, cv = 20).fit(X_train_scaled, y_train)

y_pred_lasso = model_lasso.predict(X_test_scaled)
test_error_lasso = mean_squared_error(y_test, y_pred_lasso)

print("LASSO Regression Test error:", test_error_lasso)
print("LASSO Regression Best alpha:", model_lasso.alpha_)
```

### (iv) Fit the ridge regression on the training set, with the penalization parameter chosen by leaveone-out cross-validation. Report the test error obtained.

```python
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import LeaveOneOut

ridge_alphas = np.linspace(0.01, 60, 1000)

model_ridge = RidgeCV(alphas=ridge_alphas, cv=LeaveOneOut().get_n_splits(X_train_scaled)).fit(X_train_scaled, y_train)

y_pred_ridge = model_ridge.predict(X_test_scaled)
test_error_ridge = mean_squared_error(y_test, y_pred_ridge)

print("Ridge Regression Test error:", test_error_ridge)
print("Ridge Regression Best alpha:", model_ridge.alpha_)
```

### (v) Which of the three models do you prefer? Is there much difference among the test errors?<br>

I would choose the Ridge model becasue it has the smallest test error. 

The LASSO and Ridge models have smaller error than than of the OLS model, which indicate a better performance.
